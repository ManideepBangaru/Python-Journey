## Merge

```python
# creating a data frame
ddf1 = pd.DataFrame({"Name" : ["Manideep","Shravani","Manoj","Likitha"], "PhysicsMarks" : [50, 60, 70, 80]})
ddf2 = pd.DataFrame({"Name" : ["Shravani","Manoj","Manideep","puji"], "MathsMarks" : [80, 70, 90, 100]})
```

```python
ddf1
```

```python
ddf2
```

```python
ddf3 = pd.merge(ddf1,ddf2,on="Name",how = "outer")
ddf3
```

### Group by

```python
ddf3 = pd.DataFrame({"Name" : ["Manideep","Shravani","Manoj","Likitha","Manideep","Shravani","Manoj","Likitha"], "subject": ["phy","phy","phy","phy","Mat","Mat","Mat","Mat"],"Marks" : [50, 60, 70, 80, 50, 60, 70, 80]})
```

```python
ddf3
```

```python
summary = ddf3.groupby(["Name"])["Marks"].sum().reset_index()
summary
```

### sort values

```python
summary.sort_values(by = "Marks", ascending=False, inplace = True)
summary.reset_index(drop=True,inplace=True)
summary
```

```python
flag = []
for i in range(len(summary)):
    if summary["Marks"].iloc[i] < 130:
        flag.append(1)
    else:
        flag.append(0)
```

```python
summary["Newflag"] = flag
```
